```python
df_p2_all = df_erw[df_erw["phase"] == 2]   #get phase 2 only
df_p2_all

df_p2     = df_p2_all[df_p2_all["amp"]>=1.0].index.values   #get only active mjo

df_p2
```

```python


ind_df = []  #list for the dates
phase_df = []
amp_df = []

for i in df_p2:      #loop for every phase 2 of the MJO
#     print(i)
#     p = df["phase"].iloc[i]     #Get the index of the event
    
    dt = 0
    while dt <= 100 : #placeholder, can be anything high enough
        
        print(df_erw["yyyymmdd"].iloc[i])
        pex  = df_erw["phase"].iloc[i+dt]      #get phase at t = i+dt
        pex2 = df_erw["phase"].iloc[i+dt+1]    #get phase at t = i+dt+1
            
        temp = []                          #make empty list
        temp_p = []
        temp_a = []
        if pex == 2 and pex2 == 2:         #if phase t+dt equal to 2 and t+dt+1 equal to 2, stop the loop.
            break
        elif (pex != pex2) or pex>7:       #else if phase t+dt not equal to t+dt+1 or phase t+dt > 5, go to the next cond
            if (pex2 != pex+1) or pex2 < pex:  #if phase t+dt+1 not equal to (phase t+dt)+1 or phase t+dt+1 < phase t+dt, break the loop
                break
        else:
            t = 0 #else, go for the next loop
            
            while t <= dt: #for every time, until last dt, insert the dates into the temp list
                temp.append(df_erw["yyyymmdd"].iloc[i+t])
                temp_p.append(df_erw["phase"].iloc[i+t])
                temp_a.append(df_erw["amp"].iloc[i+t])
                t += 1
        
            
            if (df_erw["phase"].iloc[i+dt] == 7)  : #if the last time is phase 7, insert it into ind_df
                ind_df.append(temp)
                phase_df.append(temp_p)
                amp_df.append(temp_a)
        dt += 1


print(ind_df)

    
```
